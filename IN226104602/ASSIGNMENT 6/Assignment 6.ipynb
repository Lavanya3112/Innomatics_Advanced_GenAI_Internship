```python
# ==========================================
# TASK 6: FUNCTION-BASED PROBLEM SOLVING
# ==========================================

# ------------------------------------------
# Problem 1: Smart Parking Lot Management
# ------------------------------------------
def parking_system(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1
        
        # Track peak usage
        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: Full (Over Capacity!)")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)


# ------------------------------------------
# Problem 2: Food Delivery Time Estimator
# ------------------------------------------
def delivery_time(distance, traffic, weather):
    # Base time (assume 5 min per km)
    time = distance * 5

    # Traffic impact
    if traffic.lower() == "high":
        time += 15
    elif traffic.lower() == "medium":
        time += 10

    # Weather impact
    if weather.lower() == "rainy":
        time += 10
    elif weather.lower() == "stormy":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


# ------------------------------------------
# Problem 3: Movie Theatre Seat Occupancy
# ------------------------------------------
def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print(f"Occupancy: {occupancy:.2f}%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")


# ------------------------------------------
# Problem 4: Cloud Server Load Classification
# ------------------------------------------
def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)
    print(f"Average CPU Load: {avg:.2f}%")

    if avg < 50:
        print("Server Status: Normal")
    elif 50 <= avg < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# ------------------------------------------
# Problem 5: Smart Classroom Resource Monitor
# ------------------------------------------
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:  # threshold
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused) if overused else "None")

    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")


# ------------------------------------------
# Problem 6: Event Registration System
# ------------------------------------------
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# ==========================================
# SAMPLE TEST RUNS (You can keep/remove)
# ==========================================

if __name__ == "__main__":
    print("\n--- Parking System ---")
    parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])

    print("\n--- Delivery Time ---")
    delivery_time(8, "High", "Rainy")

    print("\n--- Seat Occupancy ---")
    seat_occupancy(200, [1]*150)

    print("\n--- Server Load ---")
    server_load([45, 60, 70, 85, 90])

    print("\n--- Resource Monitor ---")
    resource_monitor({
        "Projector": 6,
        "AC": 9,
        "Lights": 4
    })

    print("\n--- Event Registration ---")
    event_registration(100, 105)
```
