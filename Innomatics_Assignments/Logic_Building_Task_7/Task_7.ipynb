Problem 1: Smart Light Controller

```python
class LightSystem:

    def __init__(self,roomName):
        self.roomName=roomName
        self.currentState="OFF"   # default when created

    def changeState(self,mode):

        mode=mode.upper()

        if mode=="ON":
            self.currentState="ON"
        else:
            self.currentState="OFF"   # anything else turns it off for now

        print(self.roomName,"->",self.currentState)


light1=LightSystem("Bedroom Light")
light1.changeState("ON")

# could add toggle feature later maybe
```

Problem 2: Employee ID Card System

```python
class StaffMember:

    def __init__(self,name,code,department):
        self.name=name
        self.code=code
        self.department=department

    def printCard(self):
        print("Employee Name :",self.name)
        print("Employee Code :",self.code)
        print("Department    :",self.department)


empObj=StaffMember("Rahul","EMP102","IT")
empObj.printCard()
```

Problem 3: Mobile Contact Record

```python
class PhoneContact:

    def __init__(self,name,number):
        self.name=name
        self.number=str(number)   # forcing string just in case

    def display(self):
        print("Contact Saved ✔")
        print("Name :",self.name)
        print("Phone:",self.number)


contactObj=PhoneContact("Anita",9876543210)
contactObj.display()

```

Problem 4: Product Price Tag Generator

```python
class Item:

    def __init__(self,itemName,priceValue):
        self.itemName=itemName
        self.priceValue=priceValue

    def showPrice(self):
        print("Product ->",self.itemName)
        print("MRP     -> ₹"+str(self.priceValue))


prod=Item("Headphones",2499)
prod.showPrice()
```

Problem 5: Movie Rating Display System

```python
class Film:

    def __init__(self,title,rating):
        self.title=title
        self.rating=rating

    def displayInfo(self):
        print("Now Showing:",self.title)
        print("Rating:",self.rating,"out of 5")


movieObj=Film("Inception",4.8)
movieObj.displayInfo()

```

Problem 6: Delivery Address Manager

```python
class AddressBook:

    def __init__(self,customerName,addressLine):
        self.customerName=customerName
        self.addressLine=addressLine

    def showDetails(self):
        print("Customer :",self.customerName)
        print("Location :",self.addressLine)


deliveryObj=AddressBook("Suman","Hyderabad")
deliveryObj.showDetails()
```
