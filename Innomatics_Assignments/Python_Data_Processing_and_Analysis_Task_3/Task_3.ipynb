**1: Employee Performance Bonus Eligibility**

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Identify the highest performance score
highest_score = max(employees.values())

# Find all employees who have the highest score (handle ties)
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

**2: Search Query Keyword Analysis**

```python
# Input search query entered by the user
query = "Buy mobile phone buy phone online"

# Converted to lowercase
query = query.lower()

# Split the sentence into words
words = query.split()

# Dictionary to store frequency of each word
count = {}

# Count how many times each word appears
for word in words:
    if word in count:
        count[word] = count[word] + 1
    else:
        count[word] = 1

# Dictionary to store only repeated keywords
result = {}

# Store words that appear more than once
for word in count:
    if count[word] > 1:
        result[word] = count[word]

print(result)
```

**3: Sensor Data Validation**

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid sensor readings (hour, value)
valid = []

# Variable to track hour index
i = 0

# Loop through each sensor reading
for value in sensor_readings:
    # Check if the reading is an even number (valid)
    if value % 2 == 0:
        valid.append((i, value))
    # Move to the next hour
    i = i + 1

print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

**4: Email Domain Usage Analysis**

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store count of users for each email domain
domains = {}

# Loop through each email ID
for email in emails:
    # Split email to get domain name
    parts = email.split("@")
    domain = parts[1]

    # Count how many times each domain appears
    if domain in domains:
        domains[domain] = domains[domain] + 1
    else:
        domains[domain] = 1

# Total number of users
total = len(emails)

# Calculate and display percentage usage of each domain
for domain in domains:
    percentage = (domains[domain] * 100) // total
    print(domain + ":", str(percentage) + "%")
```

**5: Sales Spike Detection**



```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Variable to store total sales
total = 0

# Calculate total sales
for value in sales:
    total = total + value

# Calculate average daily sales
average = total / len(sales)

# Variable to track day number
day = 1

# Identify and days where sales are much higher than average
for value in sales:
    if value > average * 1.2:
        print("Day", day, ":", value)
    day = day + 1
```

**6: Duplicate User ID Detection**

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store count of each user ID
count = {}

# Count how many times each user ID appears
for user in user_ids:
    if user in count:
        count[user] = count[user] + 1
    else:
        count[user] = 1

# Display only duplicate user IDs
for user in count:
    if count[user] > 1:
        print(user, "→", count[user], "times")
```
