**1: Mobile Recharge Validation System**

```python
# Function to validate recharge amount
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    # Check minimum amount
    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False

    # Check if amount is in valid plans
    if amount not in valid_plans:
        print("Invalid plan. Available plans are:", valid_plans)
        return False

    # If all conditions are satisfied
    print("Recharge successful for ₹", amount)
    return True


# Retry loop
while True:
    amount = int(input("Enter recharge amount: "))

    # Call validation function
    if validate_recharge(amount):
        break   # exit loop if recharge is valid
    else:
        print("Please try again.\n")
```

**2: Inventory Reorder Alert System**

```python
# Function to check stock status
def check_inventory(products):
    # Loop through each product in the dictionary
    for item in products:
        quantity = products[item]

        # Check if stock is below reorder level
        if quantity < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")


# Input dictionary
products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Salt": 30
}

# Call the function
check_inventory(products)
```

**3: Student Result Processing System**

```python
# Function to calculate result
def process_result(marks):

    total = 0

    # Calculate total using loop
    for mark in marks:
        total = total + mark

    average = total / len(marks)

    print("Average Marks:", average)

    # Check pass or fail
    if average >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")


# Sample marks
student_marks = [60, 55, 45, 70]

process_result(student_marks)
```

**4: Cab Fare Estimator with Retry Option**

```python
# Function to calculate cab fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_charge = 12

    # Calculate basic fare
    fare = base_fare + (distance * per_km_charge)

    # Add peak hour charges if applicable
    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


# Retry loop
while True:
    distance = int(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)

    print("Total Cab Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

**5: Employee Attendance Eligibility**

```python
# Function to check attendance eligibility
def check_attendance(attendance):
    present_count = 0

    # Count number of present days
    for day in attendance:
        if day == "P":
            present_count = present_count + 1

    # Calculate attendance percentage
    percentage = (present_count / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    # Check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Sample attendance list
attendance_list = ["P", "P", "A", "P", "P", "A"]

# Call the function and print result
result = check_attendance(attendance_list)
print("Status:", result)
```

**6: Password Strength Checker**

```python
# Function to check password strength
def check_password(password):

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum length is 8)"

    # Loop through each character
    for ch in password:
        if ch.isdigit():
            has_digit = True

        if ch in special_chars:
            has_special = True

    # Check conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"


# Sample input
user_password = input("Enter your password: ")

# Call function
result = check_password(user_password)
print(result)
```
