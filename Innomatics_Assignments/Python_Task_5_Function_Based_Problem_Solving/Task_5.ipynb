**1: Social Media – Post Engagement Analyzer**

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example
analyze_engagement([200, 300, 250, 300])
```

**3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_checker(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)


# Example
rainfall_checker([70, 75, 80, 63], 70)
```

**4: Social Media – Duplicate Account Detection**

```python
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)


# Example
detect_duplicates(["user1", "user2", "user1"])
```

**6: Agriculture – Premium Crop Price Filter**

```python
def premium_crops(prices):
    premium_list = []

    for price in prices:
        if price > 2000:
            premium_list.append(price)

    print("Premium Crops:", premium_list)


# Example
premium_crops([1500, 2500, 1800, 3200])
```

**7: System Monitoring – Application Health Checker**

```python
def check_system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)


# Example
check_system_health(7)
```

**9: E-Learning – Student Attendance Eligibility System**

```python
def attendance_checker(attendance_list):
    present = 0

    for day in attendance_list:
        if day == "P":
            present += 1

    percentage = (present / len(attendance_list)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


# Example
attendance_checker(["P", "P", "A", "P", "P"])
```

**10: Smart Electricity Bill Analyzer**

```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Status:", usage)


# Example
electricity_bill(250)
```
