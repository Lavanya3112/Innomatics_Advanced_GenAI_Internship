```python
print("2. PASS / FAIL ANALYZER\n")
# Analyze student marks to count pass and fail students

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:   # A student is considered pass if marks are 50 or above
        pass_count += 1
        print(f"You're Pass")
    else:
        fail_count += 1
        print(f"You're Fail")

print("\nTotal Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

```python
print("3: Simple Data Cleaner\n")
# Clean user names by removing extra spaces and converting to lowercase

names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

print("Cleaned Names List:", cleaned_names) # Stores standardized names in a new list
```

```python
print("4: Message Length Analyzer\n")
# Calculate the length of each message

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    print("Message:", message)
    print("Length:", len(message))

    if len(message) > 10:  # Flags messages that are longer than 10 characters

        print("Status: Long Message\n")
    else:
        print("Status: Short Message\n")
```

```python
print("5: Error Message Detector\n")
# Detect and count ERROR entries from system logs

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1  # Displays the total number of errors found

print("Total ERROR entries:", error_count)
```
